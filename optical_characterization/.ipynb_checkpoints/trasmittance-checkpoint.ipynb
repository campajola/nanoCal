```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter, FormatStrFormatter
```

### loading files

```python
ej200_t = pd.read_table(r'EJ200_T.wav', skiprows=8,sep=',', names=['t', 't_err'])
ej200_t['wl'] = np.arange(190, len(ej200_t)+190, 1)
```

```python
nanoc_t = pd.read_table(r'NANOC_T.wav', skiprows=8,sep=',', names=['t', 't_err'])
nanoc_t['wl'] = np.arange(190, len(nanoc_t)+190, 1)
```

```python
ej200_a = pd.read_table(r'EJ200.wav', sep='\t', names=['wl', 'a', 'a_err'])
#ej200_a['wl'] = np.arange(190, len(ej200_a)+190, 1)
```

```python
nanoc_a = pd.read_table(r'NANOC.wav', skiprows=8,sep=',', names=['a', 'a_err'])
nanoc_a['wl'] = np.arange(190, len(nanoc_a)+190, 1)
```

```python
ej200_a
```

```python
fig = plt.figure(figsize=(7, 5))
plt.plot(ej200_t.wl, ej200_t.t, label='EJ200')
plt.plot(nanoc_t.wl, nanoc_t.t, label='nanoC')
plt.xlabel('wavelenght [nm]', size=16)
plt.ylabel('Transmittance [%]', size=16)
plt.grid()
plt.legend(fontsize=14)
```

```python
fig = plt.figure(figsize=(7, 5))
plt.plot(ej200_t.wl, ej200_t.t, label='EJ200')
plt.plot(nanoc_t.wl, nanoc_t.t, label='nanoC')
plt.xlabel('wavelenght [nm]', size=16)
plt.ylabel('Trasmittance [%]', size=16)
plt.grid()
plt.legend(fontsize=14)
plt.xlim(400, 600)
```

```python
fig = plt.figure(figsize=(7, 5))
plt.plot(ej200_a.wl, ej200_a.a, label='EJ200')
#plt.plot(ej200_a.wl, -np.log10(nanoc_t.t/100), label='manual')
plt.plot(nanoc_a.wl, nanoc_a.a, label='nanoC')
plt.xlabel('wavelenght [nm]', size=16)
plt.ylabel('Absorbance', size=16)
plt.grid()
plt.legend(fontsize=14)
```

```python
fig = plt.figure(figsize=(7, 5))
plt.plot(ej200_a.wl, ej200_a.a, label='EJ200')
#plt.plot(ej200_a.wl, -np.log10(nanoc_t.t/100), label='manual')
plt.plot(nanoc_a.wl, nanoc_a.a, label='nanoC')
plt.xlabel('wavelenght [nm]', size=16)
plt.ylabel('Absorbance', size=16)
plt.grid()
plt.legend(fontsize=14)
plt.xlim(400, 600)
```

```python
fig = plt.figure(figsize=(7, 5))
plt.plot(nanoc_t.wl, -np.log(nanoc_t.t/100)/4, label='nanoC')
plt.plot(ej200_t.wl, -np.log(ej200_t.t/100)/4, label='EJ200')
plt.xlabel('wavelenght [nm]', size=16)
plt.ylabel('absorption coefficient (mm-1)', size=16)
plt.grid()
plt.legend(fontsize=14)
plt.yscale('log')
```

```python
fig = plt.figure(figsize=(7, 5))
plt.plot(nanoc_t.wl, -1/(np.log(nanoc_t.t/100)/4), label='nanoC')
plt.plot(ej200_t.wl, -1/(np.log(ej200_t.t/100)/4), label='EJ200')
plt.xlabel('wavelenght [nm]', size=16)
plt.ylabel('attenuation lenght (mm)', size=16)
plt.grid()
plt.legend(fontsize=14)
plt.yscale('log')
```
